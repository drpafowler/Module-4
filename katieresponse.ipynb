```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

```

```python
mean_A = 5.2
ci_A = (4.8, 5.6)
mean_B = 4.9
ci_B = (4.5, 5.3)
n = 25000

std_A = (ci_A[1] - ci_A[0]) / (2 * 1.96)
std_B = (ci_B[1] - ci_B[0]) / (2 * 1.96)

sample_A = np.random.normal(mean_A, std_A, n)
sample_B = np.random.normal(mean_B, std_B, n)

print(f"Sample A: mean={np.mean(sample_A)}, std={np.std(sample_A)}")
print(f"Sample B: mean={np.mean(sample_B)}, std={np.std(sample_B)}")
print(f"Confidence Interval for Sample A: {ci_A}")
print(f"Confidence Interval for Sample B: {ci_B}")
```

```python
t_stat, p_value = ttest_ind(sample_A, sample_B)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value:.10f}")

if p_value < 0.05:
    print("There is a statistically significant difference.")
else:
    print("There is no statistically significant difference.")
```

```python
sns.kdeplot(sample_A, label='Sample A', fill=True)
sns.kdeplot(sample_B, label='Sample B', fill=True)
plt.legend()
plt.title('KDE Plot of Sample A and Sample B')
plt.xlabel('Value')
plt.ylabel('Density')
plt.axvline(x=ci_A[0], color='blue', linestyle='--', label='CI A Lower')
plt.axvline(x=ci_A[1], color='blue', linestyle='--', label='CI A Upper')
plt.axvline(x=ci_B[0], color='orange', linestyle='--', label='CI B Lower')
plt.axvline(x=ci_B[1], color='orange', linestyle='--', label='CI B Upper')
plt.text(0.05, 0.95, f'T-statistic: {t_stat:.2f}', transform=plt.gca().transAxes, verticalalignment='top')
plt.text(0.05, 0.90, f'P-value: {p_value:.10f}', transform=plt.gca().transAxes, verticalalignment='top')
plt.show()
```
